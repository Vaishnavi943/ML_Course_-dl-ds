# **Vectors**

Vector operations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

plotting a vector

```python
plt.quiver(0,0,4,5, color='g')
plt.show()
```

```python
plt.quiver(0,0,5,4, scale_units='xy', scale=1, angles='xy')
plt.xlim(-8,10)
plt.ylim(-8,10)
plt.show()
```

```python
# multiple vector
plt.quiver(0,0,5,4, scale_units='xy', scale=1, angles='xy')
plt.quiver(0,0,-6,-2, color='g', scale_units='xy', scale=1, angles='xy')
plt.xlim(-8,8)
plt.ylim(-8,10)
plt.show()
```

addition of two vectors

```python
vector_1 = np.array([5,4])
vector_2 = np.array([6,-2])
vector_sum = vector_1 + vector_2
vector_sum
```

```python
plt.quiver(0,0,5,4, scale_units='xy', scale=1, angles='xy')
plt.quiver(0,0,6,-2, color='g', scale_units='xy', scale=1, angles='xy')
plt.quiver(0,0,11,2, color='r', scale_units='xy', scale=1, angles='xy')
plt.xlim(-8,12)
plt.ylim(-8,10)
plt.show()
```

subtraction of two vector

```python
vector_1 = np.array([2,3])
vector_2 = np.array([3,-2])
vector_sub = vector_1 - vector_2
vector_sub
```

```python
plt.quiver(0,0,5,4, scale_units='xy', scale=1, angles='xy')
plt.quiver(0,0,6,-2, color='g', scale_units='xy', scale=1, angles='xy')
plt.quiver(0,0,-1,5, color='r', scale_units='xy', scale=1, angles='xy')
plt.xlim(-8,12)
plt.ylim(-8,10)
plt.show()
```

multiply a vector with scaler

```python
vector_1 = np.array([2,3])
vector_2 = -0.5*vector_1
vector_2
```

```python
plt.quiver(0,0,2,3, scale_units='xy', scale=1, angles='xy')
plt.quiver(0,0,-1,-1.5, color='r', scale_units='xy', scale=1, angles='xy')
plt.xlim(-8,12)
plt.ylim(-8,10)
plt.show()
```

dot product of 2 vector

```python
a = np.array([2, 3])
b = np.array([3, -1])
a_dot_b = np.dot(a,b)
a_dot_b
```

cross product of 2 vector

```python
a = np.array([2, 3, 2])
b = np.array([3, -1, 2])
a_cross_b = np.cross(a,b)
a_cross_b
```

projection of "a" vector on "v" vector

```python
a = np.array([2, 5])
v = np.array([8, -6])

# magnitude of "v" vector
magnitude_v = np.sqrt(sum(v**2))

# projection of a on v
proj_a_on_v = (np.dot(a,v)/magnitude_v**2)*v

proj_a_on_v
```
